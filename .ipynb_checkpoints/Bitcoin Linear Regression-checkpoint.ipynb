```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (20.0, 10.0)

data = pd.read_csv(r"C:\Users\Elank\Documents\Backup laptop\Semester 6\Pengantar Big Data\Dataset\Bitcoin.csv")
print(data)
data.head()
```

```python
X = data['Close'].values
Y = data['Market Cap'].values
```

```python
mean_x = np.mean(X)
mean_y = np.mean(Y)

m = len(X)

numer = 0
denom = 0
for i in range(m):
    numer += (X[i] - mean_x) * (Y[i] - mean_y)
    denom += (X[i] - mean_x) ** 2
b1 = numer / denom
b0 = mean_y - (b1 * mean_x)

print(b1, b0)
```

```python
max_x = np.max(X) + 100
min_x = np.min(X) - 100

x = np.linspace(min_x, max_x, 1000)
y = b0 + b1 * x

plt.plot(x, y, color='#58b970', label='Regression Line')

plt.scatter(X, Y, color='#ef5423', label='Scatter Plot')

plt.xlabel('Close')
plt.ylabel('Market Cap')
plt.legend()
plt.show()
```
